## Classification Task dengan Random Forest

### General ML Model Training


- proses training selalu diawali dengan training set yang didalamnya terdapat X_train dan y_train.
- training set ini akan digunakan untuk model machine learning, dan setelah di training namanya trained model,
- train model ini akan melakukan prediksi terhadap nilai feature baru, dan meghasilkan nilai prediksi.

### Esemble learning : heterogeneuous & homogeneous


- sebuah teknik dalam machine learning dengan menggabungkan beberapa model.
- selalu di awali dengan traning set, dan membuat beberapa model
- dari setiap hasil model ini membuat trained model.
- lalu memasukan train model baru.
- setiap training model tersebut menghasilkan prediksi,
- setiap prediksi perlu di satukan (maturity voting) menggunakan nilai mean/rata rata atau menggunakan mode/kemunculan nilai terbanyak.

### Bagging : Bootstrap Aggregating


- menggabungkan model machine learning yang sejenis.
- karena menggunakan training model yang sama akan sia sia, karena akan menghasilkan trained model yang sama persisi, maka dari itu perlu bootstrap aggregating (Bagging)

Bootstrap aggregating : menggunakan random sampling with repalacement.
- training set yang baru disebut dengan Bag. bag akan dibuat sesuai dengan banyaknya model yang akan dibuat.
- setiap bag akan melakukan trained model ke masing masing model.
- setiap trained model ini akan membuat nilai feature yang baru.
- setelah itu semua trained model akan disatukan (maturity voting) menggunakan nilai mean/rata rata atau menggunakan mode/kemunculan nilai terbanyak.
- lalu mengahasilkan y_pred atau prediksi final.

### Random Forest


random forest : menggunakan bagging + features randomness.
- Random forest menggunakan beberapa decision tree.
- setiap model decision tree akan di trained menggunakan bag. 
- setiap bag akan memilih nilai feature dan target yang berbeda yang dipilih secara acak. 
- setiap trained model akan disatukan dan meghasilkan prediksi.

### Dataset

```python
from sklearn.datasets import load_iris
X, y = load_iris(return_X_y = True)
print(f'Dimensi Feature : {X.shape}')
print(f'Class: {set(y)}')
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,
                                             y,
                                             test_size = 0.3,
                                             random_state = 0)
```

### Classification dengan `RandomForestClassifier`

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(n_estimators = 100, random_state = 0)

model.fit(X_train, y_train)
```

membuat : `RandomForestClassifier`
- melakukan import model `from sklearn.ensemble import RandomForestClassifier`
- membuat model `model = RandomForestClassifier(n_estimators = 100, random_state = 0)`, estimators digunakan untuk mentukan berapa banyak decision tree yang akan dibangun.
- melakukan traning model menggunakan `fit` dengan `X_train` dan `y_train`


### Evaluasi Model

```python
from sklearn.metrics import classification_report

y_pred = model.predict(X_test)
print(classification_report(y_test, y_pred))
```
